## 1.2 Create Classes and Function

### 🔧 Utility Functions & Dataset Classes

The following cells define all reusable building blocks: image preprocessing functions, PyTorch `Dataset` wrappers, and the model initializer. Defining these up front keeps the training pipeline clean and modular.

```python
# ─── Reproducibility: Seed Control ────────────────────────────────────────────
# Deep learning experiments involve many sources of randomness: weight
# initialization, data shuffling, dropout, and even GPU operations. Fixing all
# random seeds ensures results are reproducible across runs.

def seed_everything(seed=23):
    """
    Fix all sources of randomness to ensure fully reproducible training.
    
    Args:
        seed (int): The integer seed to apply to all random number generators.
    """
    assert isinstance(seed, int), 'seed has to be an integer'
    
    os.environ['PYTHONHASHSEED'] = str(seed)  # Python hash randomization (dict ordering)
    random.seed(seed)                          # Python built-in random module
    np.random.seed(seed)                       # NumPy random operations
    torch.manual_seed(seed)                    # PyTorch CPU operations
    torch.cuda.manual_seed(seed)              # PyTorch GPU operations
    torch.backends.cudnn.deterministic = True  # Force deterministic cuDNN algorithms
    torch.backends.cudnn.benchmark = False     # Disable auto-tuning (trades speed for reproducibility)
```

```python
# ─── Master Image Preprocessing Function ──────────────────────────────────────
# Retinal fundus images require specialized preprocessing because:
#   1. They often have black borders (artifact of the camera lens)
#   2. The region of interest is circular (the retina disc)
#   3. Subtle vessel and lesion features have low contrast against the retinal background
#
# The preprocessing pipeline:
#   crop_black → (optional random crop) → resize → contrast enhance → circle_crop → tensor

image_size = 256  # All images will be resized to 256×256 pixels

def prepare_image(path, sigmaX=10, do_random_crop=False):
    """
    Load and preprocess a single retinal fundus image.

    Steps:
        1. Load image and convert BGR → RGB (OpenCV default is BGR)
        2. Remove black borders via crop_black()
        3. Optionally apply a random spatial crop for data augmentation
        4. Resize to a fixed square resolution
        5. Apply Gaussian illumination correction (Ben Graham's method):
               output = 4 * original − 4 * GaussianBlur(original) + 128
           This sharpens local contrast and reduces global illumination variation,
           making subtle lesions (microaneurysms, exudates) more visible.
        6. Apply circular mask to remove camera-edge artifacts
        7. Convert to PyTorch tensor with channels in (C, H, W) order

    Args:
        path (str): File path to the retinal image.
        sigmaX (int): Gaussian blur sigma for contrast enhancement. Higher values
                      smooth more, creating stronger local contrast enhancement.
        do_random_crop (bool): Whether to apply random spatial crop (training only).

    Returns:
        torch.Tensor: Preprocessed image tensor of shape (3, image_size, image_size).
    """
    # Load image and convert color space
    image = cv2.imread(path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Remove black borders from camera lens vignetting
    image = crop_black(image, tol=7)
    
    # Optional: random crop for augmentation during training
    if do_random_crop:
        image = random_crop(image, size=(0.9, 1))
    
    # Resize to uniform resolution for the CNN
    image = cv2.resize(image, (int(image_size), int(image_size)))
    
    # Ben Graham illumination correction:
    # Subtracts blurred version to remove low-frequency lighting variation,
    # then adds 128 to center pixel values around gray (avoids negative values).
    image = cv2.addWeighted(image, 4, cv2.GaussianBlur(image, (0, 0), sigmaX), -4, 128)
    
    # Circular mask: zeroes out pixels outside the retinal disc
    image = circle_crop(image, sigmaX=sigmaX)
    
    # Convert numpy (H, W, C) → PyTorch tensor → permute to (C, H, W)
    image = torch.tensor(image)
    image = image.permute(2, 1, 0)
    return image
```

```python
# ─── Black Border Cropping ─────────────────────────────────────────────────────
# Fundus cameras often produce images with large black borders. These uninformative
# regions waste model capacity and can cause the CNN to learn spurious features.
# This function uses a pixel intensity threshold to locate the actual retinal
# content and crops tightly around it.

def crop_black(img, tol=7):
    """
    Remove black border regions from a retinal fundus image.
    
    Uses a luminance threshold to create a binary mask identifying non-black pixels,
    then crops to the bounding box of the non-black region using NumPy's ix_ indexing.

    Args:
        img (np.ndarray): Input image, either grayscale (H×W) or RGB (H×W×3).
        tol (int): Pixel intensity threshold below which a pixel is considered 'black'.
                   Typical value: 7 (handles slight noise in dark borders).

    Returns:
        np.ndarray: Tightly cropped image with black borders removed.
                    Returns original image if cropping would produce an empty result.
    """
    if img.ndim == 2:
        # Grayscale: directly threshold pixel values
        mask = img > tol
        return img[np.ix_(mask.any(1), mask.any(0))]
    
    elif img.ndim == 3:
        # RGB: convert to grayscale first for the mask, then apply to all channels
        gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        mask = gray_img > tol
        
        # Safety check: if mask eliminates everything, return original image
        check_shape = img[:, :, 0][np.ix_(mask.any(1), mask.any(0))].shape[0]
        if check_shape == 0:
            return img
        else:
            # Apply the same spatial mask independently to each RGB channel
            img1 = img[:, :, 0][np.ix_(mask.any(1), mask.any(0))]
            img2 = img[:, :, 1][np.ix_(mask.any(1), mask.any(0))]
            img3 = img[:, :, 2][np.ix_(mask.any(1), mask.any(0))]
            img = np.stack([img1, img2, img3], axis=-1)
            return img
```

```python
# ─── Circular Mask (Retinal Disc Isolation) ───────────────────────────────────
# Fundus photographs capture a circular window of the retina. Pixels outside
# this circle are camera artifacts, not retinal tissue. Masking them out:
#   - Removes distracting edge artifacts
#   - Forces the CNN to focus on the clinically relevant region
#   - Standardizes images regardless of original aspect ratio

def circle_crop(img, sigmaX=10):
    """
    Apply a circular crop centered on the retinal disc.

    Steps:
        1. Resize to a square (using the largest dimension) to ensure the circle
           fits within the image boundaries
        2. Compute center coordinates and radius as half the smaller dimension
        3. Draw a filled white circle on a black mask
        4. Apply the mask via bitwise AND — only pixels inside the circle are kept

    Args:
        img (np.ndarray): Input image array of shape (H, W, C).
        sigmaX (int): Retained as parameter for API consistency with prepare_image.

    Returns:
        np.ndarray: Image with pixels outside the retinal circle set to zero (black).
    """
    height, width, depth = img.shape
    
    # Pad to square so the inscribed circle reaches the edges of the shorter dimension
    largest_side = np.max((height, width))
    img = cv2.resize(img, (largest_side, largest_side))
    height, width, depth = img.shape
    
    # Compute circle center and radius
    x = int(width / 2)
    y = int(height / 2)
    r = np.amin((x, y))  # Use the smaller half-dimension to keep circle within bounds
    
    # Create binary circular mask
    circle_img = np.zeros((height, width), np.uint8)
    cv2.circle(circle_img, (x, y), int(r), 1, thickness=-1)  # thickness=-1 fills the circle
    
    # Zero out everything outside the circle using bitwise AND with the mask
    img = cv2.bitwise_and(img, img, mask=circle_img)
    return img
```

```python
# ─── Random Spatial Crop (Training Augmentation) ─────────────────────────────
# Random cropping simulates slightly different camera positioning and zoom levels,
# encouraging the model to be robust to the precise scale and position of the retina.
# A crop between 90%–100% of the image retains most retinal content while introducing
# enough variation to be useful.

def random_crop(img, size=(0.9, 1)):
    """
    Apply a random spatial crop to an image.

    Randomly selects a crop that retains between `size[0]` and `size[1]` fraction
    of the original height and width, placed at a random position.

    Args:
        img (np.ndarray): Input image array of shape (H, W, C).
        size (tuple): (min_fraction, max_fraction) of original dimensions to retain.

    Returns:
        np.ndarray: Cropped image. Note: the returned image is NOT resized — 
                    resizing happens downstream in prepare_image().
    """
    height, width, depth = img.shape
    
    # Fraction of the image to remove (randomly chosen within the allowed range)
    cut = 1 - random.uniform(size[0], size[1])
    
    # Random top-left corner of the crop window
    i = random.randint(0, int(cut * height))  # Top offset
    j = random.randint(0, int(cut * width))   # Left offset
    
    # Bottom-right corner derived from top-left + retained size
    h = i + int((1 - cut) * height)
    w = j + int((1 - cut) * width)
    
    return img[i:h, j:w, :]
```

```python
# ─── EyeData: Custom PyTorch Dataset (Full Preprocessing Pipeline) ────────────
# PyTorch's Dataset API requires implementing __len__() and __getitem__() to allow
# DataLoader to batch, shuffle, and parallelise data loading automatically.
#
# EyeData applies the FULL preprocessing pipeline defined in prepare_image(),
# including contrast enhancement and Gaussian-based illumination correction.
# This is the primary dataset class used for both training and inference.

class EyeData(Dataset):
    
    def __init__(self, data, directory, transform=None, do_random_crop=True, itype='.png'):
        """
        Args:
            data (pd.DataFrame): DataFrame with columns ['id_code', 'diagnosis'].
            directory (str): Root folder containing the retinal images.
            transform: torchvision transforms pipeline applied after prepare_image().
            do_random_crop (bool): Enable random crop augmentation (True for training).
            itype (str): Image file extension ('.png' for 2019 data, '.jpeg' for 2015).
        """
        self.data           = data
        self.directory      = directory
        self.transform      = transform
        self.do_random_crop = do_random_crop
        self.itype          = itype

    def __len__(self):
        # Returns the number of samples — required by DataLoader
        return len(self.data)

    def __getitem__(self, idx):
        # Build full file path from id_code + extension
        img_name = os.path.join(self.directory, self.data.loc[idx, 'id_code'] + self.itype)
        
        # Apply full preprocessing pipeline (contrast enhancement, circular crop, etc.)
        image = prepare_image(img_name, do_random_crop=self.do_random_crop)
        
        # Apply torchvision transforms (ToPILImage, random flips, ToTensor, etc.)
        image = self.transform(image)
        
        # Return image and integer diagnosis label as a dict
        label = torch.tensor(self.data.loc[idx, 'diagnosis'])
        return {'image': image, 'label': label}
```

```python
# ─── Data: Simplified Dataset Class (Without Contrast Enhancement) ────────────
# This alternative Dataset class applies a lighter preprocessing pipeline:
# only black border removal, resizing, and circular masking — omitting the
# Gaussian illumination correction used in EyeData.
#
# Used primarily for exploratory visualization (Section 2.2, 3.2) where we want
# to display images without the aggressive contrast modification, allowing visual
# inspection of the raw retinal photographs as they appear clinically.

class Data(Dataset):
    
    def __init__(self, data, directory, transform=None, do_random_crop=True, itype='.png'):
        """
        Args:
            data (pd.DataFrame): DataFrame with columns ['id_code', 'diagnosis'].
            directory (str): Root folder containing the retinal images.
            transform: torchvision transforms pipeline.
            do_random_crop (bool): Retained for API consistency (not used in this class).
            itype (str): Image file extension.
        """
        self.data           = data
        self.directory      = directory
        self.transform      = transform
        self.do_random_crop = do_random_crop
        self.itype          = itype

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        img_name = os.path.join(self.directory, self.data.loc[idx, 'id_code'] + self.itype)
        
        # Load image and convert BGR → RGB
        image = cv2.imread(img_name)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        # Crop black borders → resize → circular mask (no Gaussian enhancement)
        image = crop_black(image, tol=7)
        image = cv2.resize(image, (int(image_size), int(image_size)))
        image = circle_crop(image, sigmaX=10)
        
        # Convert to tensor (H, W, C) → permute → (C, H, W)
        image = torch.tensor(image)
        image = image.permute(2, 1, 0)
        image = self.transform(image)
        
        label = torch.tensor(self.data.loc[idx, 'diagnosis'])
        return {'image': image, 'label': label}
```

```python
# ─── Model Initialization (Fine-Tuning Phase) ────────────────────────────────
# This function loads EfficientNet-B7 with weights from Phase 1 (2015 pre-training)
# and optionally freezes early layers for fine-tuning on the 2019 dataset.
#
# Layer freezing strategy:
#   - Frozen layers: gradients are not computed → faster training, prevents
#     catastrophic forgetting of learned retinal features
#   - Trainable layers (trn_layers=2): only the last 2 layer groups are updated,
#     allowing high-level feature adaptation while keeping low-level features stable

def init_model(train=True, trn_layers=2, model_name='enet_b7'):
    """
    Load and configure EfficientNet-B7 for fine-tuning on 2019 data.

    Args:
        train (bool): If True, loads in training mode with partial layer freezing.
                      If False, loads in inference mode with all layers frozen.
        trn_layers (int): Number of final layer groups to keep trainable (default: 2).
        model_name (str): Name used for locating the saved 2015 pre-trained weights.

    Returns:
        EfficientNet: Configured model ready for training or inference.
    """
    if train:
        # Load architecture + replace classification head with 5-class output
        model = EfficientNet.from_pretrained('efficientnet-b7', num_classes=5)
        
        # Load weights saved after pre-training on 2015 data
        model.load_state_dict(torch.load(
            '../input/diabetic-retinopathy-pre-training/models/model_{}.bin'.format(model_name, 1)
        ))
        
        # Freeze all layers except the last `trn_layers` groups
        # list(model.children()) returns top-level modules; [:-trn_layers] = early layers
        for child in list(model.children())[:-trn_layers]:
            for param in child.parameters():
                param.requires_grad = False  # No gradient → no update

    if not train:
        # Inference mode: load full model with all layers frozen
        model = EfficientNet.from_pretrained('efficientnet-b7', num_classes=5)
        model.load_state_dict(torch.load(
            '../input/diabetic-retinopathy-pre-training/models/model_{}.bin'.format(model_name, 1)
        ))
        for param in model.parameters():
            param.requires_grad = False  # Freeze everything for pure inference

    return model
```

## 2.2 Explore Data

### 📊 Why Explore Class Distribution?

Diabetic retinopathy severity is graded on a 5-point scale (0 = No DR, 1 = Mild, 2 = Moderate, 3 = Severe, 4 = Proliferative DR). Real-world medical datasets are often **heavily imbalanced** — the majority of patients have no or mild disease. Understanding this distribution is critical because:

- Class imbalance can bias the model toward the majority class
- It informs whether oversampling, undersampling, or weighted loss functions are needed

## 2.3  Preprocess and Augment Training Data

### 🔄 Data Augmentation Strategy

Augmentation artificially expands the training set by applying random but realistic transformations to each image. For retinal fundus images:

- **Random Rotation (±360°)**: The eye can be photographed at any angle — rotation invariance is clinically meaningful
- **Horizontal & Vertical Flips**: Lesion patterns in DR are not laterally asymmetric; flipping does not change the diagnosis
- **No color jitter**: Retinal color (red channel intensity, vessel contrast) carries diagnostic information and should not be distorted

Validation/test images receive **no augmentation** — only normalization — to get consistent, reproducible predictions.

```python
# ─── Define Data Augmentation Transforms ──────────────────────────────────────
# torchvision.transforms provides composable image transformation pipelines.
# Training and validation use DIFFERENT transforms:
#   - Training: augmentation applied to artificially expand effective dataset size
#   - Validation/Test: no augmentation (deterministic, for consistent evaluation)
#
# All transforms operate on PIL Images, so ToPILImage() must come first.

batch_size = 16   # Number of images processed per gradient update
image_size = 256  # Spatial resolution fed to EfficientNet-B7

# ── Training Transforms ────────────────────────────────────────────────────────
train_trans = transforms.Compose([
    transforms.ToPILImage(),                          # Tensor/ndarray → PIL
    transforms.RandomRotation((-360, 360)),           # Rotate by any angle (retinas have no canonical orientation)
    transforms.RandomHorizontalFlip(),                # Mirror left/right (DR lesions are not laterally asymmetric)
    transforms.RandomVerticalFlip(),                  # Mirror top/bottom
    transforms.ToTensor()                             # PIL → float tensor, normalizes to [0, 1]
])

# ── Validation Transforms (no augmentation) ────────────────────────────────────
valid_trans = transforms.Compose([
    transforms.ToPILImage(),
    transforms.ToTensor()                             # Only convert; no random transforms
])

# Test uses the same pipeline as validation
test_trans = valid_trans
```

## 2.4 - Using EFfienctNet for Transfer Learning
<center><img src="https://1.bp.blogspot.com/-DjZT_TLYZok/XO3BYqpxCJI/AAAAAAAAEKM/BvV53klXaTUuQHCkOXZZGywRMdU9v9T_wCLcBGAs/s1600/image2.png" width="1400" height="500x"></center>

### 🧠 EfficientNet-B7: Architecture Overview

EfficientNet (Tan & Le, 2019) uses a **compound scaling** method that uniformly scales network depth, width, and input resolution using a fixed set of coefficients. Unlike prior CNNs that scale only one dimension, EfficientNet achieves better accuracy with fewer parameters.

**EfficientNet-B7** is the largest variant, designed for high-resolution inputs. It is pre-trained on ImageNet (1.2M images, 1000 classes), giving it a strong foundation of visual features that transfer well to medical imaging tasks.

**Transfer Learning approach used here:**
1. Load EfficientNet-B7 with pre-trained ImageNet weights
2. Replace the final classification head with a 5-class output layer (DR grades 0–4)
3. For pre-training (Phase 1): train all parameters
4. For fine-tuning (Phase 2): freeze early layers, train only the final layers

<a name='2-4-1'></a>
### 2.4.1 - Setup Model and Choose Hyperparameters

```python
# ─── EfficientNet-B7 Architecture for Pre-Training ────────────────────────────
# In Phase 1 (pre-training on 2015 data), ALL model parameters are trainable.
# We load EfficientNet-B7 with ImageNet weights, replacing the 1000-class head
# with a 5-class head matching the DR severity grades (0–4).
#
# from_pretrained() initializes the feature extraction layers with ImageNet weights,
# then randomly initializes the new 5-class final layer. Training the whole network
# adapts these features specifically to retinal pathology.

model_name = 'enet_b7'

def init_pre_model(train=True):
    """
    Initialize EfficientNet-B7 for pre-training on the 2015 DR dataset.

    Args:
        train (bool): If True, all parameters trainable (pre-training mode).
                      If False, all parameters frozen (inference mode).

    Returns:
        EfficientNet: Model configured for the given mode.
    """
    if train:
        # Load ImageNet-pretrained EfficientNet-B7 and replace head with 5-class output
        model = EfficientNet.from_pretrained('efficientnet-b7', num_classes=5)
        # All parameters are trainable by default — no additional freezing needed

    if not train:
        # For inference: initialize architecture without pretrained weights
        model = EfficientNet.from_name('efficientnet-b7')
        # Replace the classification head with a 5-class layer
        model._fc = nn.Linear(model._fc.in_features, 5)
        # Freeze all parameters — no gradients needed for inference
        for param in model.parameters():
            param.requires_grad = False

    return model


# Inspect the full model architecture
model = init_pre_model()
print(model)
```

```python
# ─── Initialize Out-of-Fold (OOF) Prediction Placeholder ─────────────────────
# During pre-training, we track predicted class probabilities on the 2019
# validation set (5 columns = one score per DR grade).
# These predictions will be used for final performance evaluation after training.

oof_preds = np.zeros((len(test), 5))  # Shape: (num_2019_images, 5_classes)

cv_start = time.time()  # Track total training wall-clock time
```

```python
# ─── Training Hyperparameters ─────────────────────────────────────────────────
# These settings control the optimization behavior during pre-training.
#
# CrossEntropyLoss: combines LogSoftmax + NLLLoss in one step. Appropriate for
#   multi-class classification with mutually exclusive classes. Internally computes:
#   loss = -log(softmax(logit[true_class]))
#
# Adam optimizer (configured later): adaptive learning rate algorithm that maintains
#   per-parameter learning rates based on first and second gradient moments.
#   Well-suited for this task due to sparse gradients in large networks.
#
# StepLR scheduler: reduces learning rate by `gamma` every `step` epochs.
#   e.g., at epoch 5: lr = 1e-3 * 0.5 = 5e-4
#         at epoch 10: lr = 5e-4 * 0.5 = 2.5e-4
#   This enables coarse optimization early and precise fine-tuning later.

criterion = nn.CrossEntropyLoss()  # Multi-class classification loss

# Training duration and stopping criteria
max_epochs = 15  # Maximum number of passes through the training data
early_stop = 5   # Stop if no improvement after this many epochs (prevents overfitting)

# Learning rate (how large each parameter update step is)
eta = 1e-3  # Initial learning rate: 0.001

# Learning rate schedule (StepLR)
step  = 5    # Reduce LR every 5 epochs
gamma = 0.5  # Multiply LR by 0.5 at each step
```

<a name='2-4-2'></a>
## 2.4.2 - Train Model on Diabetic Retinopathy 2015 data

### ⚙️ Training Loop Mechanics

The training loop follows the standard PyTorch pattern:

1. **Forward pass**: Input batch → model → raw logits (unnormalized scores per class)
2. **Loss computation**: `CrossEntropyLoss` combines `LogSoftmax` + negative log-likelihood, appropriate for multi-class classification
3. **Backward pass**: `loss.backward()` computes gradients via automatic differentiation
4. **Parameter update**: `optimizer.step()` applies the Adam update rule
5. **Validation**: Model is set to `eval()` mode, gradients disabled (`torch.no_grad()`), predictions collected

**Adam optimizer** adapts the learning rate per parameter using first and second moment estimates — well-suited for noisy, sparse gradients in vision tasks.

**StepLR scheduler** reduces the learning rate by `gamma` every `step` epochs, enabling coarse learning early on and fine-grained convergence later.

```python
# ─── Prepare DataLoaders for Pre-Training ─────────────────────────────────────
# DataLoader wraps the Dataset to:
#   - Batch samples together into tensors of shape (batch_size, C, H, W)
#   - Shuffle training data each epoch (prevents learning order-dependent patterns)
#   - Parallelize data loading across CPU workers (num_workers=4)
#
# Pre-training configuration:
#   - Training set: full 2015 dataset (35,126 images) with augmentation
#   - Validation set: 2019 APTOS dataset (3,662 images), NO augmentation
#     This cross-dataset validation tests domain generalization ability.

# Use all 2015 data for training
data_train = train
data_valid = test  # 2019 images serve as the validation domain

# Create datasets with appropriate transforms
train_dataset = EyeData(
    data=data_train,
    directory='../input/diabetic-retinopathy-resized/resized_train/resized_train',
    transform=train_trans,
    itype='.jpeg'
)
valid_dataset = EyeData(
    data=data_valid,
    directory='../input/aptos2019-blindness-detection/train_images',
    transform=valid_trans,
    itype='.png'
)

# DataLoaders: training shuffled, validation ordered (reproducible evaluation)
train_loader = torch.utils.data.DataLoader(
    train_dataset, batch_size=batch_size, shuffle=True,  num_workers=4
)
valid_loader = torch.utils.data.DataLoader(
    valid_dataset, batch_size=batch_size, shuffle=False, num_workers=4
)
```

```python
# ─── Pre-Training Loop (2015 Dataset) ─────────────────────────────────────────
# This is the core training loop. Each epoch consists of two phases:
#   1. TRAINING: forward pass → compute loss → backpropagate → update weights
#   2. VALIDATION: forward pass only (no gradients) → compute loss and kappa
#
# Early stopping halts training when validation kappa stops improving,
# preventing overfitting to the large 2015 dataset.

# Metric trackers across epochs
val_kappas = []
val_losses = []
trn_losses = []
bad_epochs = 0  # Counts consecutive epochs without improvement

# Initialize model and move to GPU/CPU
model = init_pre_model()
model = model.to(device)

# Adam optimizer: updates ALL model parameters (pre-training = full model training)
optimizer = optim.Adam(model.parameters(), lr=eta)

# StepLR scheduler: halves LR every `step` epochs
scheduler = lr_scheduler.StepLR(optimizer, step_size=step, gamma=gamma)

for epoch in range(max_epochs):

    # ── Setup ──────────────────────────────────────────────────────────────────
    epoch_start = time.time()
    trn_loss = 0.0
    val_loss = 0.0
    fold_preds = np.zeros((len(data_valid), 5))  # 5 class probabilities per image

    # ── Training Phase ─────────────────────────────────────────────────────────
    model.train()  # Enables dropout and batch normalization in training mode

    for batch_i, data in enumerate(train_loader):
        inputs = data['image']
        labels = data['label'].view(-1)                        # Flatten to 1D
        inputs = inputs.to(device, dtype=torch.float)
        labels = labels.to(device, dtype=torch.long)
        
        optimizer.zero_grad()                                   # Clear previous gradients

        with torch.set_grad_enabled(True):
            preds = model(inputs)                               # Forward pass: logits
            loss  = criterion(preds, labels)                    # CrossEntropy loss
            loss.backward()                                     # Backpropagate gradients
            optimizer.step()                                    # Update parameters

        trn_loss += loss.item() * inputs.size(0)               # Accumulate batch loss

    # ── Validation / Inference Phase ───────────────────────────────────────────
    model.eval()   # Disables dropout; uses running stats for batch normalization

    for batch_i, data in enumerate(valid_loader):
        inputs = data['image']
        labels = data['label'].view(-1)
        inputs = inputs.to(device, dtype=torch.float)
        labels = labels.to(device, dtype=torch.long)

        with torch.set_grad_enabled(False):  # No gradient computation = faster inference
            preds = model(inputs).detach()
            # Store raw logits (argmax later gives predicted class)
            fold_preds[batch_i * batch_size:(batch_i + 1) * batch_size, :] = preds.cpu().numpy()

        loss      = criterion(preds, labels)
        val_loss += loss.item() * inputs.size(0)

    # Save validation predictions for this epoch
    oof_preds = fold_preds

    # Advance the learning rate scheduler
    scheduler.step()

    # ── Evaluation ─────────────────────────────────────────────────────────────
    fold_preds_round = fold_preds.argmax(axis=1)  # Predicted class = argmax of logits
    val_kappa = metrics.cohen_kappa_score(
        data_valid['diagnosis'], fold_preds_round.astype('int'), weights='quadratic'
    )

    val_kappas.append(val_kappa)
    val_losses.append(val_loss / len(data_valid))
    trn_losses.append(trn_loss / len(data_train))

    # ── Early Stopping Check ────────────────────────────────────────────────────
    print('- epoch {}/{} | lr = {} | trn_loss = {:.4f} | val_loss = {:.4f} | val_kappa = {:.4f} | {:.2f} min'.format(
        epoch + 1, max_epochs, scheduler.get_lr()[-1],
        trn_loss / len(data_train), val_loss / len(data_valid), val_kappa,
        (time.time() - epoch_start) / 60
    ))

    if epoch > 0:
        if val_kappas[epoch] < val_kappas[epoch - bad_epochs - 1]:
            bad_epochs += 1  # No improvement this epoch
        else:
            bad_epochs = 0   # Reset counter on improvement
```

## 2.4.3 - Evaluation

### 📈 Evaluation Metrics

**Cohen's Quadratic Kappa (QWK)** is the primary metric for this task. It measures agreement between predicted and true labels, accounting for ordinal distance between classes. A prediction of grade 2 when the truth is grade 4 is penalized more than a prediction of grade 3.

- Kappa = 1.0: Perfect agreement
- Kappa = 0.0: Random agreement
- Kappa < 0: Worse than random

QWK is preferred over accuracy here because the target variable is **ordinal** (grade 0 < 1 < 2 < 3 < 4), and clinical severity ordering matters.

```python
# ─── Training Dynamics: Loss and Kappa Curves ────────────────────────────────
# Plotting training vs. validation loss across epochs provides a diagnostic view
# of the learning process:
#   - Both curves decreasing: model is learning effectively
#   - Training loss low, validation loss rising: overfitting (model memorizing training data)
#   - Both curves high and flat: underfitting (model capacity too low, or LR too high)
#
# The Kappa curve shows whether ordinal grade accuracy is improving per epoch.

sns.set()
fig = plt.figure(figsize=(15, 5))

# Loss curves
plt.subplot(1, 2, 1)
plt.plot(trn_losses, 'red',   label='Training Loss')
plt.plot(val_losses, 'green', label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Cross-Entropy Loss')
plt.title('Loss Dynamics — Pre-Training (2015 → 2019)')
plt.legend()

# Quadratic Weighted Kappa curve
plt.subplot(1, 2, 2)
plt.plot(val_kappas, 'blue', label='Validation QWK')
plt.xlabel('Epoch')
plt.ylabel('Cohen's Quadratic Weighted Kappa')
plt.title('Kappa Dynamics — Pre-Training')
plt.legend()
```

```python
# ─── Threshold-Based Rounding and Final Performance (2015 Pre-Training) ───────
# The model outputs logits → argmax gives integer class predictions (0–4).
# In ordinal classification problems, applying custom thresholds can improve
# performance by accounting for class imbalance and borderline predictions.
#
# Here, predictions are binned using thresholds [0.5, 1.5, 2.5, 3.5]:
#   pred < 0.5          → grade 0
#   0.5 ≤ pred < 1.5    → grade 1
#   1.5 ≤ pred < 2.5    → grade 2
#   2.5 ≤ pred < 3.5    → grade 3
#   pred ≥ 3.5          → grade 4
# These thresholds are fixed here but could be optimized on a held-out set.

oof_preds_round = oof_preds.argmax(axis=1)  # Initial integer predictions from argmax

# Apply custom thresholds (ordinal binning)
coef = [0.5, 1.5, 2.5, 3.5]
for i, pred in enumerate(oof_preds_round):
    if pred < coef[0]:
        oof_preds_round[i] = 0
    elif pred >= coef[0] and pred < coef[1]:
        oof_preds_round[i] = 1
    elif pred >= coef[1] and pred < coef[2]:
        oof_preds_round[i] = 2
    elif pred >= coef[2] and pred < coef[3]:
        oof_preds_round[i] = 3
    else:
        oof_preds_round[i] = 4

# Compute final loss and quadratic weighted kappa
oof_loss  = criterion(torch.tensor(oof_preds), torch.tensor(test['diagnosis']).view(-1).type(torch.long))
oof_kappa = metrics.cohen_kappa_score(test['diagnosis'], oof_preds_round.astype('int'), weights='quadratic')
print('OOF loss  = {:.4f}'.format(oof_loss))
print('OOF kappa = {:.4f}'.format(oof_kappa))
```

```python
# ─── Normalized Confusion Matrix (2015 Pre-Training Evaluation) ───────────────
# The confusion matrix shows, for each true DR grade (rows), what grade the
# model predicted (columns). Normalizing by row gives per-class recall rates.
#
# Ideal matrix: high values on the diagonal (correct predictions) and near-zero
# off-diagonal values. Common errors to look for:
#   - Grade 0 confused with Grade 1 (mild vs. no DR — clinically acceptable)
#   - Grade 4 confused with lower grades (proliferative DR missed — clinically dangerous)
#
# The seaborn heatmap makes systematic confusion patterns easy to spot visually.

# Build row-normalized confusion matrix
cm = confusion_matrix(test['diagnosis'], oof_preds_round)
cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]  # Normalize to recall per class
annot = np.around(cm, 2)  # Round to 2 decimals for readability

# Plot
fig, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(cm, cmap='Blues', annot=annot, lw=0.5,
            xticklabels=['Grade 0', 'Grade 1', 'Grade 2', 'Grade 3', 'Grade 4'],
            yticklabels=['Grade 0', 'Grade 1', 'Grade 2', 'Grade 3', 'Grade 4'])
ax.set_xlabel('Predicted Grade')
ax.set_ylabel('True Grade')
ax.set_title('Normalized Confusion Matrix — Pre-Training (2015 → 2019)')
ax.set_aspect('equal')
```

```python
# ─── Per-Class Classification Report (2015 Pre-Training) ─────────────────────
# The classification report provides Precision, Recall, and F1-score per class:
#
#   Precision: Of all images predicted as grade X, what fraction truly are grade X?
#              High precision = few false positives for that class.
#   Recall:    Of all images truly grade X, what fraction were predicted as grade X?
#              High recall = few missed cases of that class.
#   F1-score:  Harmonic mean of precision and recall — balanced metric.
#
# For clinical DR screening, RECALL for grades 3 and 4 is most critical —
# missing severe DR has serious consequences for patient vision.

from sklearn.metrics import classification_report

print('\n Classification Report (2015 Pre-Training Evaluation):\n')
print(classification_report(test['diagnosis'], oof_preds_round))
```

## 3.4 - Load Pre-Trained EFfienctNetB7 trained on 2015 data

### 🔁 Transfer Learning: Loading the Pre-Trained Weights

Rather than initializing from random weights or from ImageNet alone, we load the weights saved after Phase 1 (pre-training on the 2015 data). This gives us a model already specialized in recognizing retinal pathology patterns.

Only the **final fully-connected layer** (`model._fc`) is trained in this phase — all earlier feature extraction layers remain frozen. This is effective because:
- The retinal features learned from 35K images generalize well to the 3.6K 2019 images
- Training only the small FC head dramatically reduces the risk of overfitting
- Training is much faster with fewer parameters to update

<a name ="3-4-1"></a>
### 3.4.1 Setup Model and Choose Hyperparameters

```python
# ─── Initialize Fine-Tuning Model (Load 2015 Pre-Trained Weights) ────────────
# This call to init_model() loads EfficientNet-B7 with weights from Phase 1.
# Early layers are frozen; only the last 2 layer groups remain trainable.
# The full architecture is printed to verify the model structure.

model_name = 'enet_b7'

# Load model in training mode (partial layer freezing applied inside init_model)
model = init_model(model_name=model_name)
print(model)
```

```python
# ─── Stratified K-Fold Cross-Validation Setup ────────────────────────────────
# K-Fold CV splits the 2019 training data into `num_folds` groups.
# STRATIFIED K-Fold ensures each fold has approximately the same class proportion
# as the full dataset — critical when classes are imbalanced (grades 3 and 4 are rare).
#
# With 4 folds:
#   - Each fold uses 75% of data for training and 25% for validation
#   - All images get a validation prediction exactly once (no leakage)
#   - Final OOF kappa estimates performance on the full training dataset

from sklearn.model_selection import KFold, StratifiedKFold

num_folds = 4  # Number of CV folds (balance between compute cost and estimate reliability)

# Create stratified fold indices, shuffled for randomness
skf    = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=seed)
splits = list(skf.split(train['id_code'], train['diagnosis']))

# Placeholder for out-of-fold predictions (one per training image)
oof_preds = np.zeros((len(train), 1))

cv_start = time.time()  # Track total cross-validation time
```

```python
# ─── Fine-Tuning Hyperparameters ──────────────────────────────────────────────
# Same hyperparameter setup as pre-training.
# Key note: in the CV loop (cell below), the Adam optimizer is configured to
# only update model._fc (the final layer), NOT all parameters as in pre-training.
# This reflects the frozen-backbone fine-tuning strategy.

criterion = nn.CrossEntropyLoss()  # Multi-class classification loss

max_epochs = 15  # Maximum epochs per fold
early_stop = 5   # Patience: stop after 5 non-improving epochs

eta   = 1e-3  # Initial learning rate
step  = 5     # LR decay step (epochs)
gamma = 0.5   # LR decay factor
```

<a name = "3-4-2"></a>
## 3.4.2 Train Model on Diabetic Retinopathy 2019 dat

```python
# ─── Stratified K-Fold Cross-Validation Training Loop ────────────────────────
# This is the main fine-tuning loop. For each fold:
#   1. Split the 2019 training data into train/validation subsets
#   2. Load the 2015 pre-trained model (fresh for each fold)
#   3. Freeze early layers; train only the final FC layer
#   4. Run the training/validation epoch loop with early stopping
#   5. Save the best model checkpoint (by validation kappa)
#   6. Store OOF predictions from the best epoch

for fold in tqdm(range(num_folds)):

    # ── Data Preparation ───────────────────────────────────────────────────────
    print('-' * 30)
    print('FOLD {}/{}'.format(fold + 1, num_folds))
    print('-' * 30)

    # Split indices for this fold (stratified by diagnosis grade)
    data_train = train.iloc[splits[fold][0]].reset_index(drop=True)
    data_valid = train.iloc[splits[fold][1]].reset_index(drop=True)

    # Create Dataset objects with fold-specific subsets
    train_dataset = EyeData(
        data=data_train,
        directory='../input/aptos2019-blindness-detection/train_images',
        transform=train_trans,
        itype='.png'
    )
    valid_dataset = EyeData(
        data=data_valid,
        directory='../input/aptos2019-blindness-detection/train_images',
        transform=valid_trans,
        itype='.png'
    )

    # DataLoaders for this fold
    train_loader = torch.utils.data.DataLoader(
        train_dataset, batch_size=batch_size, shuffle=True,  num_workers=4
    )
    valid_loader = torch.utils.data.DataLoader(
        valid_dataset, batch_size=batch_size, shuffle=False, num_workers=4
    )

    # ── Model Preparation ──────────────────────────────────────────────────────
    val_kappas = []
    val_losses = []
    trn_losses = []
    bad_epochs = 0

    # Restore best OOF predictions from previous fold (if fold > 0)
    if fold > 0:
        oof_preds = oof_preds_best.copy()

    # Load fresh copy of pre-trained model for each fold (avoids fold-to-fold contamination)
    model = init_model(train=True)
    model = model.to(device)

    # Only train the final classification layer — all other layers are frozen in init_model()
    optimizer = optim.Adam(model._fc.parameters(), lr=eta)
    scheduler = lr_scheduler.StepLR(optimizer, step_size=step, gamma=gamma)

    # ── Epoch Loop ────────────────────────────────────────────────────────────
    for epoch in range(max_epochs):
        epoch_start = time.time()
        trn_loss    = 0.0
        val_loss    = 0.0
        fold_preds  = np.zeros((len(data_valid), 1))

        # ── Training ──────────────────────────────────────────────────────────
        model.train()

        for batch_i, data in enumerate(train_loader):
            inputs = data['image']
            labels = data['label'].view(-1)
            inputs = inputs.to(device, dtype=torch.float)
            labels = labels.to(device, dtype=torch.long)
            optimizer.zero_grad()

            with torch.set_grad_enabled(True):
                preds = model(inputs)              # Forward pass
                loss  = criterion(preds, labels)   # Compute loss
                loss.backward()                    # Backpropagate
                optimizer.step()                   # Update FC layer weights only

            trn_loss += loss.item() * inputs.size(0)

        # ── Validation ────────────────────────────────────────────────────────
        model.eval()

        for batch_i, data in enumerate(valid_loader):
            inputs = data['image']
            labels = data['label'].view(-1)
            inputs = inputs.to(device, dtype=torch.float)
            labels = labels.to(device, dtype=torch.long)

            with torch.set_grad_enabled(False):
                preds = model(inputs).detach()
                _, class_preds = preds.topk(1)    # Get top-1 predicted class
                fold_preds[batch_i * batch_size:(batch_i + 1) * batch_size, :] = class_preds.cpu().numpy()

            loss      = criterion(preds, labels)
            val_loss += loss.item() * inputs.size(0)

        # Save validation predictions to OOF array at the correct indices
        oof_preds[splits[fold][1]] = fold_preds

        scheduler.step()

        # ── Metrics ───────────────────────────────────────────────────────────
        fold_preds_round = fold_preds  # Already integer class indices (from topk)
        val_kappa = metrics.cohen_kappa_score(
            data_valid['diagnosis'], fold_preds_round.astype('int'), weights='quadratic'
        )
        val_kappas.append(val_kappa)
        val_losses.append(val_loss / len(data_valid))
        trn_losses.append(trn_loss / len(data_train))

        # ── Early Stopping & Checkpoint Saving ────────────────────────────────
        print('- epoch {}/{} | lr = {} | trn_loss = {:.4f} | val_loss = {:.4f} | val_kappa = {:.4f} | {:.2f} min'.format(
            epoch + 1, max_epochs, scheduler.get_lr()[-1],
            trn_loss / len(data_train), val_loss / len(data_valid), val_kappa,
            (time.time() - epoch_start) / 60
        ))

        if epoch > 0:
            if val_kappas[epoch] < val_kappas[epoch - bad_epochs - 1]:
                bad_epochs += 1
            else:
                bad_epochs = 0

        # Save model and OOF predictions when validation improves
        if bad_epochs == 0:
            oof_preds_best = oof_preds.copy()
            torch.save(model.state_dict(), 'models/model_{}_fold{}.bin'.format(model_name, fold + 1))

        # Trigger early stopping if patience is exhausted
        if bad_epochs == early_stop:
            print('Early stopping. Best: loss = {:.4f}, kappa = {:.4f} (epoch {})'.format(
                np.min(val_losses), val_kappas[np.argmin(val_losses)], np.argmin(val_losses) + 1
            ))
            print('')
            break

        if epoch == (max_epochs - 1):
            print('Max epochs reached. Best: loss = {:.4f}, kappa = {:.4f} (epoch {})'.format(
                np.min(val_losses), val_kappas[np.argmin(val_losses)], np.argmin(val_losses) + 1
            ))
            print('')
            break

# Use the best OOF predictions (from best epoch across all folds)
oof_preds = oof_preds_best

print('')
print('Finished in {:.2f} minutes'.format((time.time() - cv_start) / 60))
```

## 3.4.3 - Evaluation


### 📊 Final Model Evaluation

The final evaluation uses the **Out-of-Fold (OOF) predictions** — predictions made on each sample when it was part of the validation set. This gives an unbiased estimate of model performance across the entire 2019 training dataset.

Three complementary views of performance are presented:
1. **Training/validation loss curves**: Check for overfitting (diverging curves) or underfitting (both losses high)
2. **Confusion matrix (normalized)**: Shows per-class recall — which DR grades are confused with each other
3. **Classification report**: Precision, recall, and F1-score per class — useful for understanding performance on minority classes (grades 3 and 4)

```python
# ─── Training Dynamics: Loss and Kappa Curves (2019 Fine-Tuning) ─────────────
# These plots reflect the last fold's training dynamics.
# For a full picture, you would need to track and plot metrics across all folds.
# Signs of healthy training:
#   - Training and validation losses both decrease steadily
#   - Kappa increases steadily then plateaus (early stopping kicks in)
#   - No large gap between training and validation loss (no severe overfitting)

sns.set()
fig = plt.figure(figsize=(15, 5))

# Loss curves (last fold)
plt.subplot(1, 2, 1)
plt.plot(trn_losses, 'red',   label='Training Loss')
plt.plot(val_losses, 'green', label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Cross-Entropy Loss')
plt.title('Loss Dynamics — Fine-Tuning on 2019 Data (Last Fold)')
plt.legend()

# Kappa curve (last fold)
plt.subplot(1, 2, 2)
plt.plot(val_kappas, 'blue', label='Validation QWK')
plt.xlabel('Epoch')
plt.ylabel('Cohen's Quadratic Weighted Kappa')
plt.title('Kappa Dynamics — Fine-Tuning (Last Fold)')
plt.legend()
```

```python
# ─── Final OOF Performance (2019 Fine-Tuning) ────────────────────────────────
# Evaluate out-of-fold predictions against the ground truth for the full
# 2019 training dataset. Because each image was predicted exactly once
# (when it was in the validation split of its fold), this is an unbiased
# estimate of how the model generalizes to unseen data.

oof_preds_round = oof_preds.copy()  # Integer class predictions from K-Fold CV

oof_kappa = metrics.cohen_kappa_score(
    train['diagnosis'], oof_preds_round.astype('int'), weights='quadratic'
)
print('OOF Quadratic Weighted Kappa = {:.4f}'.format(oof_kappa))
```

```python
# ─── Normalized Confusion Matrix (2019 Fine-Tuning OOF) ──────────────────────
# Same interpretation as Section 2.4.3. Key things to examine:
#   - How well does the model separate Grade 0 from Grade 1? (subtle boundary)
#   - Does the model reliably detect Grade 4 (proliferative DR)? (critical for screening)
#   - Are errors concentrated in adjacent grades (clinically acceptable) or
#     do they span multiple grades (more dangerous)?

cm = confusion_matrix(train['diagnosis'], oof_preds_round)
cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]  # Row normalize → recall per class
annot = np.around(cm, 2)

fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(cm, cmap='Blues', annot=annot, lw=0.5,
            xticklabels=['Grade 0', 'Grade 1', 'Grade 2', 'Grade 3', 'Grade 4'],
            yticklabels=['Grade 0', 'Grade 1', 'Grade 2', 'Grade 3', 'Grade 4'])
ax.set_xlabel('Predicted Grade')
ax.set_ylabel('True Grade')
ax.set_title('Normalized Confusion Matrix — OOF Predictions (2019 Fine-Tuning)')
ax.set_aspect('equal')
```

```python
# ─── Per-Class Classification Report (2019 Fine-Tuning OOF) ─────────────────
# The final classification report provides a comprehensive breakdown of model
# performance across all DR severity grades.
#
# Focus areas for clinical interpretation:
#   - Grade 0 Recall: Are healthy eyes correctly identified? (affects screening efficiency)
#   - Grade 3-4 Recall: Are severe cases flagged for urgent care? (patient safety)
#   - Macro-average F1: Overall performance across all classes (ignores class imbalance)
#   - Weighted-average F1: Performance weighted by class frequency

from sklearn.metrics import classification_report

print('\n Classification Report — 2019 OOF Predictions:\n')
print(classification_report(train['diagnosis'], oof_preds_round))
```
